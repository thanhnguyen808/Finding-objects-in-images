```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

# Đọc ảnh
image_path = "Finding/1.jpg"
img = cv2.imread(image_path)

# Lấy kích thước ảnh
height, width, _ = img.shape

# Lấy màu tại tọa độ cuối cùng của ảnh (ở góc dưới bên phải)
target_color = img[height-1, width-1]

# Chuyển đổi màu sắc của target_color sang định dạng phù hợp (BGR -> RGB)
target_color_rgb = tuple(reversed(target_color))

# Tìm tất cả các điểm có màu giống target_color
# Tạo một mặt nạ (mask) với những điểm có màu giống hệt màu tại tọa độ cuối cùng
mask = np.all(img == target_color, axis=-1)

# Tìm chỉ số cột đầu tiên có màu giống target_color
split_x = None
for x in range(width):
    if np.any(mask[:, x]):
        split_x = x
        break

if split_x is None:
    print("Không tìm thấy vị trí có màu giống màu tại tọa độ cuối cùng.")
else:
    # Cắt ảnh thành hai phần
    left_part = img[:, :split_x]
    right_part = img[:, split_x:]

    # Lưu ảnh đã cắt
    cv2.imwrite("/mnt/data/left_part.jpg", left_part)
    cv2.imwrite("/mnt/data/right_part.jpg", right_part)

    print(f"Tách thành công tại vị trí {split_x}! Đã lưu 2 ảnh: left_part.jpg & right_part.jpg")

    # Hiển thị ảnh cắt bằng matplotlib
    plt.figure(figsize=(10, 5))

    # Hiển thị phần bên trái
    plt.subplot(1, 2, 1)
    plt.imshow(cv2.cvtColor(left_part, cv2.COLOR_BGR2RGB))
    plt.title("Phần bên trái")
    plt.axis('off')

    # Hiển thị phần bên phải
    plt.subplot(1, 2, 2)
    plt.imshow(cv2.cvtColor(right_part, cv2.COLOR_BGR2RGB))
    plt.title("Phần bên phải")
    plt.axis('off')

    plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

# Đọc ảnh
image_path = "Finding/1.jpg"
img = cv2.imread(image_path)

# Lấy kích thước ảnh
height, width, _ = img.shape

# Kích thước lưới (số lượng ô theo chiều ngang và dọc)
grid_size = 10

# Kích thước mỗi ô trong lưới
cell_height = height // grid_size
cell_width = width // grid_size

# Vẽ lưới và thêm số vào các tọa độ
for i in range(grid_size):
    for j in range(grid_size):
        # Tính toán tọa độ góc trên bên trái của mỗi ô
        x = j * cell_width
        y = i * cell_height
        
        # Vẽ một vòng tròn tại tọa độ (x, y)
        cv2.circle(img, (x + cell_width // 2, y + cell_height // 2), 5, (0, 0, 255), -1)
        
        # Thêm số vào các điểm tọa độ (số theo thứ tự)
        cv2.putText(img, f"({i+1},{j+1})", (x + cell_width // 2 + 5, y + cell_height // 2), 
                    cv2.FONT_HERSHEY_SIMPLEX, 0.4, (255, 255, 255), 1, cv2.LINE_AA)

# Hiển thị ảnh với lưới và số tọa độ
plt.figure(figsize=(10, 10))
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import os

# Đọc ảnh
image_path = "Finding/1.jpg"
img = cv2.imread(image_path)

# Lấy kích thước ảnh
height, width, _ = img.shape

# Kích thước lưới (số lượng ô theo chiều ngang và dọc)
grid_size = 10  # Chia ảnh thành 10x10 ô

# Kích thước mỗi ô trong lưới
cell_height = height // grid_size
cell_width = width // grid_size

# Vẽ lưới và thêm số vào các tọa độ
for i in range(grid_size):
    for j in range(grid_size):
        # Tính toán tọa độ góc trên bên trái của mỗi ô
        x = j * cell_width
        y = i * cell_height
        
        # Vẽ một vòng tròn đỏ tại tọa độ (x, y)
        cv2.circle(img, (x + cell_width // 2, y + cell_height // 2), 5, (0, 0, 255), -1)
        
        # Thêm số vào các điểm tọa độ (số theo thứ tự)
        cv2.putText(img, f"({i+1},{j+1})", (x + cell_width // 2 + 10, y + cell_height // 2 + 10), 
                    cv2.FONT_HERSHEY_SIMPLEX, 0.4, (0, 0, 0), 1, cv2.LINE_AA)

# Đảm bảo thư mục đích tồn tại
output_dir = "Split"
os.makedirs(output_dir, exist_ok=True)

# Đường dẫn file lưu ảnh
output_path = os.path.join(output_dir, "image_with_grid.jpg")

# Lưu ảnh đã chỉnh sửa
cv2.imwrite(output_path, img)

# Hiển thị ảnh với lưới và số tọa độ
plt.figure(figsize=(10, 10))
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.axis('off')  # Tắt trục tọa độ của matplotlib
plt.show()

print(f"Ảnh đã lưu tại: {output_path}")
```
