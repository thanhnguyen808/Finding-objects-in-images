```python
import cv2
import matplotlib.pyplot as plt
import os

# c nh
image_path = "Finding/1.jpg"
img = cv2.imread(image_path)

# Ly kch thc nh
height, width, _ = img.shape

# Ta  ct ( v tr x = 1400 pixels)
split_x = 1408

# Ct nh thnh 2 phn ti x = 1400
left_part = img[:, :split_x]  # Phn bn tri ca nh
right_part = img[:, split_x:]  # Phn bn phi ca nh

# Lu cc nh  ct
output_dir = "Split"
os.makedirs(output_dir, exist_ok=True)

# ng dn lu nh
left_part_path = os.path.join(output_dir, "left_part.jpg")
right_part_path = os.path.join(output_dir, "right_part.jpg")

cv2.imwrite(left_part_path, left_part)
cv2.imwrite(right_part_path, right_part)

# Hin th nh  ct
plt.figure(figsize=(10, 10))

# Hin th phn bn tri
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(left_part, cv2.COLOR_BGR2RGB))
plt.title("Left Part")
plt.axis('off')

# Hin th phn bn phi
plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(right_part, cv2.COLOR_BGR2RGB))
plt.title("Right Part")
plt.axis('off')

plt.show()

print(f"nh  tch v lu ti: {left_part_path} v {right_part_path}")
```
