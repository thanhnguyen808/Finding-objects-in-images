```python
import cv2
import matplotlib.pyplot as plt
import os

# Đọc ảnh
image_path = "Finding/1.jpg"
img = cv2.imread(image_path)

# Lấy kích thước ảnh
height, width, _ = img.shape

# Tọa độ cắt (ở vị trí x = 1400 pixels)
split_x = 1408

# Cắt ảnh thành 2 phần tại x = 1400
left_part = img[:, :split_x]  # Phần bên trái của ảnh
right_part = img[:, split_x:]  # Phần bên phải của ảnh

# Lưu các ảnh đã cắt
output_dir = "Split"
os.makedirs(output_dir, exist_ok=True)

# Đường dẫn lưu ảnh
left_part_path = os.path.join(output_dir, "left_part.jpg")
right_part_path = os.path.join(output_dir, "right_part.jpg")

cv2.imwrite(left_part_path, left_part)
cv2.imwrite(right_part_path, right_part)

# Hiển thị ảnh đã cắt
plt.figure(figsize=(10, 10))

# Hiển thị phần bên trái
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(left_part, cv2.COLOR_BGR2RGB))
plt.title("Left Part")
plt.axis('off')

# Hiển thị phần bên phải
plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(right_part, cv2.COLOR_BGR2RGB))
plt.title("Right Part")
plt.axis('off')

plt.show()

print(f"Ảnh đã tách và lưu tại: {left_part_path} và {right_part_path}")
```
